```python
## group metrics 1
fun2_2 = ['count']
fun2_2_dict = {}
col_f2_2 = ['street']
for cols in ['jams_weekend','jams_weekday','jams_morning','jams_night','jams_midnight']:
    fun2_2_dict[cols] = fun2_2
    for met in ['cnt']:
        col_f2_2.append(cols+'_'+met)
        
alert_f1 = dg1.agg(fun2_2_dict).reset_index()
alert_f1.columns = col_f2_2
```
